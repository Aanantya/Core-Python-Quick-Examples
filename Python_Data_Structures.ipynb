# Data Structures
Python data structures are a way of organizing data and accessing it as required.

```python
#List

my_list = [1, 2, 3, 4]

print(my_list[0])

my_list.pop()
print(my_list)

my_list.append(4)
print(my_list)

my_list.remove(4)
print(my_list)
```

```python
# Tuple -> they are immutable

my_tuple = (1, 2, 3, 4)

print(my_tuple)
print(my_tuple[0])
print(my_tuple[-1])
```

```python
#Set -> Do not allow duplicate elements

set1 = {1, 1, 2, 3, 4}
set2 = {3, 4}

print('set1 = ', set1)
print('set2 = ', set2)

print('\nset1 - set2 = ', (set1 - set2))
print('\nset1 | set2 = ', (set1 | set2))
print('\nset1 & set2 = ', (set1 & set2))

print('\nset2.issubset(set1) = ', set2.issubset(set1))
print('\nset1.difference(set2) = ', set1.difference(set2))
print('\nset1.union(set2) = ', set1.union(set2))
print('\nset1.intersection(set2) = ', set1.intersection(set2))
```

```python
#dict

my_dict = {1: 'a', 2: 'b', 3:'c', 4:'d'}

print(my_dict[1])
print(my_dict.keys())
print(my_dict.values())
print(my_dict.items())
```

```python
#Counter -> sub-class of dict, used to keep a count of elements

from collections import Counter

counter = Counter([1, 1, 2, 3, 4, 4, 4])
print(counter)

print(counter.most_common(1))
```

```python
#OrderedDict -> sub-class of dict, it remembers the order

from collections import OrderedDict

od = OrderedDict({4: 3, 1: 2, 2: 1, 3: 1})

od.pop(2)

od[2] = 1

print(od)
```

```python
#ChinMap -> combine multiple dict and return list of dict

from collections import ChainMap

dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}

chain = ChainMap(dict1, dict2)

print(chain)

for x in chain.items():
  print(x)
```

```python
#NamedTuple -> name every value in Tuple

from collections import namedtuple

student = namedtuple('Student', ['name', 'age'])

s = student(['John', ], [36, 20])

print(s.name, s.age)
```
