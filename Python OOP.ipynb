OOP = > Object Oriented Programming
Python is a OOP language

```python
#class and object

class MyClass:
  x = 5

obj = MyClass()
print(obj.x)
```

```python
#passing values

class MyClass():
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def show(self):
    print(self.name, self.age)

obj = MyClass("John", 36)
obj.show()
```

```python
#modify

class MyClass():
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def show(self):
    print(self.name, self.age)

obj = MyClass("John", 36)
obj.show()
obj.age = 20
obj.name = 'Sam'
obj.show()
```

**Inheritence**

```python
#child class

class ChildClass(MyClass):
  pass

obj = ChildClass('Sam', 20)
obj.show()
```

```python
#super() -> inherit all the properties of the parent class

class ChildClass(MyClass):
  def __init__(self, name, age):
    super().__init__(name, age)
    name = 'Mark'
    print(name, age)

obj = ChildClass('Sam', 20)
```

```python
#polymorphism -> class with same method name

class Addition():
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def calculate(self):
    print(self.x + self.y)

class Subtraction():
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def calculate(self):
    print(self.x - self.y)

class Multiplication():
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def calculate(self):
    print(self.x * self.y)


add = Addition(3, 4)
sub = Subtraction(5, 4)
mul = Multiplication(2, 3)

obj = [add, sub, mul]
for i in obj:
  i.calculate()
```

```python
#data hiding


class MyClass():
  __salary = 50000  #private

  def __init__(self, name, age):
    self.name = name
    self.age = age

  def show(self):
    print(self.__salary, self.name, self.age)


obj = MyClass('John', 36)
obj.show()
```

```python
print(obj.__salary) #would throw error
```

```python
print(obj._MyClass__salary) #would work
```
