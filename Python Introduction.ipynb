**Data Types**
1. str, int, float, complex
2. list, tuple, set, frozenset, dict, range
3. Bool
4. NoneType
4. Bytes

1. List = [1, 2, 3, 4] -> lists are mutable
2. Tuple = (1, 2, 3, 4) -> tuples are immutable
3. Set = {1, 2, 3, 4} -> set is mutable but it stores immutable values (no repetition allowed)



```python
#list

l = [1, 2, 3, 4, 1, 3]

print(len(l)) #length
print(l) # whole list
print(l[2]) #value by index
l.append(5) #add value
print(l)
l.remove(2) #remove value
print(l)
```

```python
#tuple

t = (1, 2, 3, 4, 1, 3)

print(len(t))
print(t)
print(t[2])
```

```python
#set

s = {1, 2,3 ,4 ,1 ,3}

print(len(s))
print(s)
s.add(5)
print(s)
s.pop()
print(s)
```

```python
#dict

d = {1: 'a', 2: 'b', 3: 'c'}

print(len(d))
print(d)
print(d[2])
print(d.keys(), d.values())
d[4] = 'd'
print(d)
```
