**File Operations**
1. Create
2. Open
3. Read
4. Write
5. Close
6. Delete

**Also**

'x' = Create, return error if file already exist

'a' = Append, create if file does not exist

'w' = Write, create if file does not exist

```python
#create file

file = open('test.txt', 'x')
```

```python
import os

os.listdir()
```

```python
file = open('test.txt', 'x')
```

```python
file = open('test.txt', 'w')
```

```python
file = open('test.txt', 'a')
```

```python
#write to file

file = open('test.txt', 'w')
file.write('Hello world!')
file.close()

# read from file
file = open('test.txt', 'r')
print(file.read())
file.close()
```

```python
#write multiple lines

file = open('test.txt', 'w')
file.writelines(['Hey how are you?\n', 'I am doing good thank you.\n'])
file.close()


#read multiple lines

file = open('test.txt', 'r')
print(file.readlines())
file.close()

#OR

file = open('test.txt', 'r')
lines = file.readlines()

for line in lines:
  print(line)

file.close()
```

```python
import os

print(os.listdir())

print(os.path.exists('test.txt'))
```

```python
#remove

os.remove('test.txt')

print(os.path.exists('test.txt'))
```
