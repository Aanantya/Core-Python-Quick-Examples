#Multiprocessing


**Simple program using Multiprocessing**

```python
import multiprocessing as mp
import time

def counter_increasing():
  i = 0
  while i < 1000000:
    time.sleep(1)
    i += 10

def counter_decreasing():
  i = 1000000
  while i > 0:
    time.sleep(1)
    i -= 10


start_time = time.time()

mp.Process(target=counter_increasing).start()
mp.Process(target=counter_decreasing).start()

end_time = time.time()

print(f"Execution time = {end_time - start_time}")
```
